```python
from pathlib import Path
import pandas as pd

DATA_DIR = Path("/home/jovyan/.cache/kagglehub/datasets/obulisainaren/multi-cancer/versions/3/Multi Cancer/Multi Cancer/ALL")   
classes = sorted([p.name for p in DATA_DIR.iterdir() if p.is_dir()])

rows = [] # Empty list called rows. 
for cls in classes: # for the different subclasses in the class loop 
    for f in (DATA_DIR/cls).rglob("*.jpg"):# for a specifc subclass folder 
        rows.append({"path": str(f), "label": cls}) # add the path and subclass of the cancer to list called 'rows'
df = pd.DataFrame(rows) # new variable called df which is a dataframe with all the value of rows with path and label 
print(df.label.value_counts(), len(df)) # prints the amount of each subclass that the cancer has 
'''
Glioma	5000 
brain_menin	5000 
brain_tumor	5000
'''
```

```python
from sklearn.model_selection import train_test_split

# This does tha datasplit for the df dataframe with the path and label 
# Training split needs to be 80/10/10 or 70/15/15 ???? 


y = df["label"] # takes the columns of label from the dataframe
# Test 15%

train_ratio = 0.70
val_ratio = 0.15
test_ratio = 0.15
# Use the train_test_split function twice when using training, validation, testing 
                                                    # 70%        15%        15% 
# Step 1: Carve out the test set (15%) from the full dataset
X_train_val, X_test, y_train_val, y_test = train_test_split(
    df, df["label"],
    test_size=test_ratio,
    random_state=1337,
    stratify=df["label"]   # Keeps class proportions balanced
)

# Step 2: Split the remaining 85% into train (70%) and val (15%)
# val_ratio / (train_ratio + val_ratio) = 0.15 / 0.85 ≈ 0.1765
X_train, X_val, y_train, y_val = train_test_split(
    X_train_val, y_train_val,
    test_size=val_ratio / (train_ratio + val_ratio),
    random_state=1337,
    stratify=y_train_val   # Keeps class proportions balanced here too
)


df_train = X_train.copy()
df_val   = X_val.copy()
df_test  = X_test.copy()

df_train["split"] = "train"
df_val["split"]   = "val"
df_test["split"]  = "test"

split_df = pd.concat([df_train, df_val, df_test], ignore_index=True)
split_df.to_csv("leukemia_cancer_split.csv", index=False)

# --- Sanity checks ---
total = len(df)
print(f"\nTotal samples : {total}")
print(f"Train         : {len(df_train)} ({len(df_train)/total:.1%})")
print(f"Val           : {len(df_val)}  ({len(df_val)/total:.1%})")
print(f"Test          : {len(df_test)}  ({len(df_test)/total:.1%})")

print("\nClass counts by split:")
print(split_df.groupby(["split", "label"]).size().unstack(fill_value=0))
```
