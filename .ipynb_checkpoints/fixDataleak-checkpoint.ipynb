```python
from pathlib import Path
import pandas as pd

DATA_DIR = Path("/home/jovyan/.cache/kagglehub/datasets/obulisainaren/multi-cancer/versions/3/Multi Cancer/Multi Cancer/Kidney Cancer")   # e.g., .../Breast Cancer
classes = sorted([p.name for p in DATA_DIR.iterdir() if p.is_dir()])

rows = []
for cls in classes:
    for f in (DATA_DIR/cls).rglob("*.jpg"):
        rows.append({"path": str(f), "label": cls})
df = pd.DataFrame(rows)
print(df.label.value_counts(), len(df))
```

```python
from PIL import Image
import imagehash

def perceptual_hash(p):
    with Image.open(p) as im:
        im = im.convert("RGB")
        return str(imagehash.phash(im))  # robust perceptual hash

df["phash"] = df["path"].apply(perceptual_hash)

# Keep one image per (label, phash) pair
df = df.sort_values("path").drop_duplicates(subset=["label","phash"]).reset_index(drop=True)
```

```python
from sklearn.model_selection import train_test_split

y = df["label"]
# Test 15%
df_trainval, df_test = train_test_split(
    df, test_size=0.15, stratify=y, random_state=1337
)
# Val 15% overall -> 0.1765 of remaining 85%
df_train, df_val = train_test_split(
    df_trainval, test_size=0.1765, stratify=df_trainval["label"], random_state=2024
)

df_train["split"]="train"; df_val["split"]="val"; df_test["split"]="test"
split_df = pd.concat([df_train, df_val, df_test], ignore_index=True)
split_df.to_csv("kidney_cancer_split.csv", index=False)

# Sanity checks
print("Class counts by split:")
print(split_df.groupby(["split","label"]).size().unstack(fill_value=0))
```

```python
import tensorflow as tf
import pandas as pd

IMG_SIZE = (256, 256)
BATCH_SIZE = 8
AUTOTUNE = tf.data.AUTOTUNE

manifest = pd.read_csv("kidney_cancer_split.csv")
label_to_int = {l:i for i,l in enumerate(sorted(manifest["label"].unique()))}

def make_ds(manifest, split):
    sub = manifest[manifest["split"]==split].copy()
    paths = sub["path"].tolist()
    labels = [label_to_int[l] for l in sub["label"]]

    ds = tf.data.Dataset.from_tensor_slices((paths, labels))

    def _load(path, label):
        img = tf.io.read_file(path)
        img = tf.io.decode_jpeg(img, channels=3)
        img = tf.image.resize(img, IMG_SIZE)
        img = tf.cast(img, tf.float32) / 255.0
        return img, tf.cast(label, tf.int32)

    if split == "train":
        ds = ds.shuffle(len(paths), seed=1337, reshuffle_each_iteration=True)

    return ds.map(_load, num_parallel_calls=AUTOTUNE).batch(BATCH_SIZE).prefetch(AUTOTUNE)

train_ds = make_ds(manifest, "train")
val_ds   = make_ds(manifest, "val")
test_ds  = make_ds(manifest, "test")
```
